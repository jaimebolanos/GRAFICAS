```python


import matplotlib.pyplot as plt
import numpy as np


t = np.linspace(0, 10, 100)


plt.plot(t, 0.5*np.cos(2*t), label='movimiento libre ')
plt.plot(t,np.exp(-2*t)+np.exp(-7*t), label='con fricción')
plt.plot(t,20*np.exp(-2*t)*np.cos(2*t)+np.sin(2*t), label='con fricción')

# Add a legend
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
