## Gráficas sencillas

La figura es la ventana o página general en la que se dibuja todo.
Es el componente de nivel superior de todos los que considerará en los siguientes puntos. 
Puedes crear múltiples figuras independientes. Una figura puede tener varias otras cosas, como un suptítulo, que es un título centrado en la figura. T. Lambién encontrará que puede agregar una leyenda y una barra de color,"por ejemplo, a su Figura.
A la figura se le añaden Ejes. Los Ejes es el área en la que se trazan los datos con funciones como plot () y scatter () y que pueden tener marcas, etiquetas, etc. asociadas. Esto explica por qué las figuras pueden contener múltiples ejes.




```python
from pylab import *

x = linspace(0, 5, 4)
y = x ** 2


figure(1,figsize = (8, 9), dpi = 20, facecolor = 'blue',edgecolor = 'red', marker='k--' )
plot(x, y, 'r',linewidth ='50' ,linestyle= '--')
xlabel('x este es x', fontsize=(29))
ylabel('y')
title('titulo')
show();
 
```

```python
from pylab import *

x = linspace(0, 5, 4)
y = x ** 2



subplot(1,2,2) #  1 fila dos columnas y dibuja en la 1
plot(x, y, 'b--')
subplot(1,2,1)# 1 fila dos columnas y dibuja en la 2
plot(y, x, 'g*-');
```

## Gráficas un poco más versatiles

```python
from pylab import *
x = linspace(0, 5, 4)
y = x ** 2


fig = plt.figure(2)

axes = fig.add_axes([1, 1, 1,0.5]) # left, bottom, width, height (range 0 to 1)

axes.plot(x, y, 'g')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('titulo novedoso');
```

```python
plt.plot(x, y, 'k--', linewidth = 2)
```
