# Gr√°ficas Usando sympy.plotting

```python
from sympy.plotting import plot
expr = x**2
expr_range = (x,-2,3)

title = '$y = {}$'.format(latex(expr)) # analizar este comando

plot(expr, expr_range, title = title, xlabel = 'x este es', ylabel = 'y');
```

```python
# Varias funciones
expr_1 = x
expr_2 = x**2
expr_3 = x**3

plot(expr_1, expr_2, expr_3, (x, -1, 1));
```

```python
# VArias funciones con diferentes rangos

expr_1 = x**2
range_1 = (x,-2,2)

expr_2 = x
range_2 = (x,-1,1)

plot(
    (expr_1,range_1),
    (expr_2,range_2)
);
```

# Varias gr√°ficas con diferentes colores

### Trazar varias gr√°ficas en colores diferentes no es sencillo

1. Colocar pase show=false para interrumpir la representaci√≥n gr√°fica y guardar el objeto en una variable
2. Indexar el objeto para obtener los serie de datos individuales
3. Mostrar la gr√°fica


```python
expr_1 = x**2
range_1 = (x,-2,2)

expr_2 = x
range_2 = (x,-1,1)

p = plot(
    (expr_1,range_1),
    (expr_2,range_2),
    show = False
);

p[0].line_color = 'r'
p[1].line_color = 'b'

p.show()
```

# Leyenda adicional a cada gr√°fica
 
1. Escribir leyend=true cuando se est√© construyendo el objeto
2. Colocar los nombres de cada seerie separadamente


```python
import pylab as P
expr_1 = x**2
range_1 = (x,-2,2)

expr_2 = x
range_2 = (x,-1,1)

p = plot(
    (expr_1,range_1),
    (expr_2,range_2),
    show = False,
    legend = True
);

p[0].line_color = 'r'
p[1].line_color = 'b'

p[0].label = 'Velocidad instantanea'
p[1].label = 'Aceleraci√≥n'


p.show()
```

# Otros ejemplos

```python
from sympy import symbols
from sympy.plotting import plot
x = symbols('x')

#Defined the function to be differentiated
f = exp(x) * sin(2*x)
#plot1 = f
p1 = f
#plot2 = f'
p2 = diff(f, x)
#P is the plot of f and f'
p = plot(p1, p2, (x, 0, 10), show=false)
#change the color of p2
p[1].line_color = 'r'

p.show()
```

# Metodo append  
append(arg)

    Adds an element from a plot‚Äôs series to an existing plot.

```python
from sympy import symbols
from sympy.plotting import plot
x = symbols('x')
p1 = plot(x*x)
p2 = plot(x)
p3=plot(x-1)
p1.append(p2[0])
p1.append(p3[0])
p1
```

# Metodo extend

```python
from sympy import symbols
from sympy.plotting import plot
x = symbols('x')
p1 = plot(x*x)
p2 = plot(x)
p1.extend(p2)
p1;
```
