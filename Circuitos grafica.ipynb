```python
import matplotlib.pyplot as plt
import numpy as np
t=np.arange(0,10,0.1)
# np.sin(x)
# np.e**x
# np.e**2 
i1=(609/101)*np.e**(-20*t)
i2=(30/101)*np.sin(2*t)-(3/101)*np.cos(2*t)
i=(609/101)*np.e**(-20*t)+(30/101)*np.sin(8*t)-(3/101)*np.cos(8*t)

plt.plot(t,i,linestyle=':', color='green')
plt.title("Gráfica del problema 7.20", fontsize='20', color='r' )
plt.xlabel('tiempo en minutos $t$',fontsize='20')
plt.ylabel('corriente en amperios A $A$')
plt.grid('true')
# plt.line_color=('green')
# plt.text(0,0.4,'la corriente comienza a oscilar')

plt.show()
```
