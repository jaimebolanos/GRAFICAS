```python
plt.isinteractive()

# Importación de matplotlib y numpy
import matplotlib.pyplot as plt
import numpy as np

# Se preparan los datos
x = np.linspace(0, 5, 100)

# Se grafican los datos
plt.plot(x, x**2+5, label='Este es un título ')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

## Uso de Estilos  con algo de programación orientada a objetos

```python
import matplotlib.pyplot as plt
plt.style.use('classic')

%matplotlib inline 

# No abre una ventana por cada gráfica
import numpy as np

x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()

ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal')
leg = ax.legend();

```

```python
import matplotlib.pyplot as plt
plt.style.use('bmh')

%matplotlib inline
import numpy as np

x = np.linspace(0, 10, 1000)
f, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
ax.axis('equal') 
leg = ax.legend();
```
